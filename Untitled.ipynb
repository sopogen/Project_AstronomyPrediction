```python
import pandas as pd
from lightgbm import LGBMClassifier
```

```python
train = pd.read_csv('data/train.csv', index_col=0)
test = pd.read_csv('data/test.csv', index_col=0)
sample_submission = pd.read_csv('data/sample_submission.csv', index_col=0)
```

```python
column_number = {}
for i, column in enumerate(sample_submission.columns):
    column_number[column] = i
    
def to_number(x, dic):
    return dic[x]

train['type_num'] = train['type'].apply(lambda x: to_number(x, column_number))
```

```python
train.iloc[:, :22]
```

```python
#Scaling

from sklearn.preprocessing import RobustScaler

rb_scaler = RobustScaler()
train.iloc[:, 1:22] = rb_scaler.fit_transform(train.iloc[:, 1:22])
test.iloc[:, :20] = rb_scaler.fit_transform(test.iloc[:, :20])

test.head()
```

```python
train_x = train.drop(columns=['type', 'type_num'], axis=1)
train_y = train['type_num']
test_x = test
```

```python
#Scaling

from sklearn.preprocessing import RobustScaler

rb_scaler = RobustScaler()
train_x[:,:] = rb_scaler.fit_transform(train_x[:,:])

train_x.head()
```

```python
from lightgbm import LGBMClassifier

lgbm =  LGBMClassifier(boosting_type='dart', class_weight=None, colsample_bytree=1.0,
        importance_type='split', learning_rate=0.05, max_depth=-1,
        min_child_samples=20, min_child_weight=0.001, min_split_gain=0.0,
        n_estimators=400, n_jobs=-1, num_leaves=127, objective=None,
        random_state=None, reg_alpha=0.0, reg_lambda=0.0, silent=True,
        subsample=1.0, subsample_for_bin=200000, subsample_freq=0)

lgbm.fit(train_x, train_y)
```

```python
y_pred = lgbm.predict_proba(test_x)
```

```python
submission = pd.DataFrame(data=y_pred, columns=sample_submission.columns, index=sample_submission.index)
submission.to_csv('submission.csv', index=True)
```
